```python
import timm

def download_pretrained_swin_transformer(model_name='swin_tiny_patch4_window7_224'):
    model = timm.create_model(model_name, pretrained=True)
    return model
```

```python
model = download_pretrained_swin_transformer()
```

```python
import os
from PIL import Image
import torch
from torchvision.transforms import Compose, Resize, ToTensor, Normalize

def classify_images(folder_path, model):
    # Define the transformations
    transform = Compose([
        Resize((224, 224)),
        ToTensor(),
        Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])

    model.eval()

    # List all files in the folder
    for filename in os.listdir(folder_path):
        # Check if the file is an image
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            # Load and transform the image
            image_path = os.path.join(folder_path, filename)
            image = Image.open(image_path).convert('RGB')
            image = transform(image).unsqueeze(0)

            # Classify the image
            with torch.no_grad():
                output = model(image)
                _, predicted = output.max(1)

            # Print the filename and the inferred class
            print(f'{filename}: {predicted.item()}')
```

```python
classify_images('data', model)
```
