```python
import torch
from pytorch_pretrained_gans import make_gan
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
```

```python
G = make_gan(gan_type='biggan')
y = G.sample_class(batch_size=1)
z = G.sample_latent(batch_size=1)
x = G(z=z, y=y)
```

```python
x_new = x[0, :, :, :]
x_new = x_new - x_new.min()
x_new = x_new / x_new.max()

# Transpose tensor to (H, W, C)
x_new = x_new.permute(1, 2, 0)

# Display image
plt.imshow(x_new.detach().numpy())
plt.show()
```
